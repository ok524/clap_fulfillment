```python
import pandas as pd
import json
```

```python
filename = "./Job Functions from websites - Sheet2 (copy).tsv"
outfile = "jobs_func_map.json"
```

```python
df = pd.read_csv(filename, sep="\t", encoding="utf8")
df
```

```python
websites = [x for x in list(set(list(df["website"])))]
websites
```

```python
df[df["website"] == "page personnel"]
```

```python
func_collection = {}
for item in websites:
# for item in ["page personnel"]:
    web = item.lower().replace(" ", "")
    func_collection[web] = {}
    
    thisweb_df = df[df["website"] == item]
    for row in range(len(thisweb_df)):
        func = list(thisweb_df["job_function"])[row]
        code = list(thisweb_df["category_id"])[row]
        name = list(thisweb_df["category_name"])[row]

        func_collection[web][func] = {
            "code": code,
            "name": name
        }
func_collection
```

```python
## Load from JSON
# with open(filename, mode="r", encoding="utf8") as f:
#     atext = f.read()
#     ajson = json.loads(atext)
```

```python
# Write from JSON
with open(outfile, mode="w", encoding="utf8") as f:
    atext = json.dumps(func_collection, ensure_ascii=False)
    f.write(atext)
```

```python
from gensim.models import Word2Vec
from gensim.models import Phrases
import pandas as pd
import re
```

```python
file_path = "/media/veracrypt1/_Foreign/_vg/clap_all/01fulfillment/notebooks/clustering_job/Job Functions from websites - Sheet1.tsv"
```

```python
df = pd.read_csv(file_path, encoding="utf8", sep="\t")
```

```python
sentences = []
sentence_arr = []
for aid, afunc in zip(df["ID"], df["job_function"]):
    # print(f"{aid} 11> {afunc}")
    newfunc = afunc.replace("&", " & ").replace(",", " ").replace("(", "  ").replace("/", " ")
    newfunc = newfunc.replace(")", "  ")
    newfunc = newfunc.replace("  ", " ").replace("  ", " ").replace("  ", " ").replace("  ", " ")
    newfunc = re.sub(r"(\w)([A-Z])[a-z]", r"\1 \2", newfunc)
    print(f"{aid} 22> {newfunc}")
    sentences.append(newfunc.split(" "))
    # sentence_arr.append(afunc.split(" "))
```

```python
sentences
```

```python
# sentences = list(df["job_function"])
# sentences = sentence_arr

model = Word2Vec(sentences, min_count=1)
```

```python
bigram_transformer = Phrases(sentences)
```

```python
model = Word2Vec(bigram_transformer[sentences], min_count=1)
```

```python
w2c = dict()
for item in model.wv.vocab:
    w2c[item] = model.wv.vocab[item].count

w2cSorted = dict(sorted(w2c.items(), key=lambda x: x[1],reverse=True))
```

```python
w2cSorted
```

```python
model.wv['Accounting']
```

```python
vb = sorted([x for x in w2cSorted.keys() if len(x) > 1], key=lambda x: len(x), reverse=True)
```

```python
jobfunc_vectors = []
for idx in range(len(sentences)):
    new_vec = None
    for w in sentences[idx]:
        if w in [x.replace("_", " ") for x in vb]:
            if new_vec is None:
                new_vec = model.wv[w]
            else:
                new_vec = new_vec + model.wv[w]
            print(w)
    jobfunc_vectors.append(new_vec)
```

```python
X = model[model.wv.vocab]
X = jobfunc_vectors
```

```python
from nltk.cluster import KMeansClusterer
import nltk
NUM_CLUSTERS=27
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25, avoid_empty_clusters=True)
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
print (assigned_clusters)
```
